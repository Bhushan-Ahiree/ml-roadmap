## Objective

Understand why Machine Learning exists, how it differs from traditional programming,
and identify when ML should or should not be used.

## Traditional Programming

```python
marks = 78

if marks >= 40:
    print("Pass")
else:
    print("Fail")
```

The programmer explicitly writes the rules.

```python
area = 1200
bedrooms = 3
near_metro = True

price = 0

if area > 1000:
    price += 500000

if bedrooms == 3:
    price += 400000

if near_metro:
    price += 300000

print(price)
```

This works for a few rules.

Imagine writing thousands of such rules for millions of houses.
It quickly becomes impossible to maintain.

This is where Machine Learning becomes useful.

## Machine Learning

Instead of writing rules manually, we provide historical data.

The learning algorithm discovers patterns and creates a model.

The trained model can then make predictions on unseen data.

## Real World Examples

- House Price Prediction
- Property Recommendation
- Swimming Pool Detection
- Spam Email Detection

## Key Takeaways

- Machine Learning learns patterns from data.
- Use ML when writing rules is difficult.
- Don't use ML for simple rule-based problems.
