# NumPy Reshape

## Objective

Learn how to reshape arrays.

```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5,6])

arr
```

```python
arr.reshape(2,3)
```

```python
arr.reshape(3,2)
```

```python
arr.reshape(2,-1)
```

```python
matrix = arr.reshape(2,3)

matrix
```

```python
matrix = matrix.flatten()
```

```python
matrix.shape
```

```python
matrix.size
```

```python
numbers = np.array([10,20,30,40,50,60,70,80])
```

```python
numbers = numbers.reshape(4,2)

numbers
```

```python
numbers = numbers.reshape(2,-1)

numbers
```

```python
numbers = numbers.flatten()

numbers
```

```python
numbers.size
```

```python
numbers.shape
```
