# Data Fundamentals

## Objective

Understand the basic terminology used in Machine Learning datasets.

```python
## Example Dataset
dataset = [
    {
        "Area": 1200,
        "Bedrooms": 2,
        "Age": 5,
        "Price": 75
    },
    {
        "Area": 1500,
        "Bedrooms": 3,
        "Age": 2,
        "Price": 95
    },
    {
        "Area": 900,
        "Bedrooms": 2,
        "Age": 10,
        "Price": 52
    }
]

for row in dataset:
    print(row)
```

## Features

Area

Bedrooms

Age

Target

Price

```python
print("Number of samples:", len(dataset))
print("First Sample")

print(dataset[0])
print("Area")

print(dataset[0]["Area"])

print("Price")

print(dataset[0]["Price"])
```


## Observation

Each dictionary represents one sample.

Each key represents a feature.

Price is the target variable.

## Key Takeaways

- A dataset is a collection of samples.
- Samples contain features.
- Features are used to predict the target.
