⭐ Biggest Beginner Mistake

Many people think this:

arr * arr

means matrix multiplication.

It doesn't.

It means:

Element-wise multiplication

```python
arr * arr
```

Without using a loop:

Add 5 bonus marks.

Multiply every mark by 2.

Square every mark.

Divide every mark by 10.

```python
marks = np.array([55,68,72,90,81])
```

```python
marks += 5
marks
```

```python
marks *= 2
marks
```

```python
marks *= marks
marks
```

```python
marks = marks // 10
marks
```
