```python
import pandas as pd
```

```python
student = {
    "Name":["Bhushan","Rahul","Sneha","Amit","Karan"],
    "Age":[22,21,23,24,22],
    "City":["Pune","Mumbai","Nashik","Delhi","Pune"],
    "Marks":[82,90,88,75,95]
}

df = pd.DataFrame(student)

df
```

```python
df[df["Marks"] > 80]  # does not modify the original dataframe, it returns a new dataframe with only those rows where Marks > 80
```

```python
df[(df["Marks"] > 90) | (df["City"] == "Mumbai")]
```
