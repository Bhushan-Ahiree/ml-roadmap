# NumPy Indexing & Slicing

## Objective

Learn how to access data from NumPy arrays.

```python
import numpy as np
```

```python
arr = np.array([10,20,30,40,50])

arr
```

```python
print(arr[0])

print(arr[2])

print(arr[-1])
```

```python
print(arr[1:4])
```

```python
print(arr[:3])

print(arr[2:])

print(arr[:])
```

```python
print(arr[::2])
```

```python
matrix = np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])

matrix
```

```python
print(matrix[0,1])

print(matrix[2,2])
```

```python
print(matrix[:,1])
```

```python
print(matrix[1,:])
```

```python
print(matrix[0:2,1:3])
```
