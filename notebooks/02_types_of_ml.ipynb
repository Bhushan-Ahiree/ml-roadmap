# Types of Machine Learning

## Objective

Understand the three major types of Machine Learning and identify when each one should be used.



## Supervised Learning

The training data contains input features and their corresponding correct outputs (labels).

Goal:
Learn the relationship between input and output so the model can predict unseen data.


```python
# Example of labeled data

properties = [
    {"Area": 1200, "Bedrooms": 2, "Price": 7500000},
    {"Area": 1500, "Bedrooms": 3, "Price": 9500000},
    {"Area": 900, "Bedrooms": 2, "Price": 5200000}
]

for property in properties:
    print(property)
```

### Examples

- House Price Prediction
- Spam Detection
- Loan Approval
- Customer Churn Prediction
## Unsupervised Learning

The training data does not contain labels.

The algorithm discovers hidden patterns or groups within the data.

```python

# Unlabeled data

properties = [
    [1200, 2],
    [1500, 3],
    [900, 2],
    [1800, 4]
]

print(properties)
```

Possible discovered groups

- Luxury Properties
- Mid-range Properties
- Budget Properties

No one tells the algorithm these groups.



## Reinforcement Learning

An agent interacts with an environment.

Good actions receive rewards.
Bad actions receive penalties.

The goal is to learn the best strategy.

```python

reward = 100
penalty = -20

print(f"Reward: {reward}")
print(f"Penalty: {penalty}")
```

## Comparison

| Type | Labels | Goal |
|------|--------|------|
| Supervised | Yes | Predict |
| Unsupervised | No | Find Patterns |
| Reinforcement | Rewards | Learn Strategy |


## Key Takeaways

- Supervised Learning uses labeled data.
- Unsupervised Learning finds hidden patterns.
- Reinforcement Learning learns through rewards and penalties.
