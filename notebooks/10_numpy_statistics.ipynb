# NumPy Statistics

## Objective

Perform common statistical operations on arrays.

```python
import numpy as np
```

```python
marks = np.array([72,85,91,65,88])

marks
```

```python
print("Sum :", marks.sum())
```

```python
print("Average :", marks.mean())
```

```python
print("Maximum :", marks.max())
```

```python
print("Minimum :", marks.min())
```

```python
print("Standard Deviation :", marks.std())
```

```python
print("Shape :", marks.shape)

print("Size :", marks.size)

print("Data Type :", marks.dtype)
```
