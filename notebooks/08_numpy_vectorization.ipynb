# NumPy Vectorization

## Objective

Learn how NumPy performs operations without loops.

```python
import numpy as np
```

```python
arr = np.array([10,20,30,40,50])

arr
```

```python
arr * 2
```

```python
arr / 10
```

```python
arr ** 2
```

```python
a = np.array([1,2,3])

b = np.array([4,5,6])

a + b
```

```python
a * b
```

```python
numbers = [1,2,3]

print(numbers * 2)
```

```python
arr = np.array([1,2,3])

print(arr * 2)
```

⭐ Biggest Beginner Mistake

Many people think this:

arr * arr

means matrix multiplication.

It doesn't.

It means:

Element-wise multiplication

```python
arr * arr
```
